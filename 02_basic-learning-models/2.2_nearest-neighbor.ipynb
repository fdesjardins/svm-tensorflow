```python
# NN using TensforFlow
import numpy as np
import tensorflow as tf

# MNIST data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
Xtr, Ytr = mnist.train.next_batch(5000) # nn candidates
Xte, Yte = mnist.test.next_batch(200) # for testing

xtr = tf.placeholder("float", [None, 784])
xte = tf.placeholder("float", [784])

# nearest neighbor using L1 distance
distance = tf.reduce_sum(tf.abs(tf.add(xtr, tf.negative(xte))), reduction_indices=1)

# prediction; get nearest neighbor
prediction = tf.arg_min(distance, 0)

accuracy = 0

init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for i in range(len(Xte)):
        nn_index = sess.run(prediction, feed_dict={xtr: Xtr, xte: Xte[i, :]})
        print("test", i, "prediction:", np.argmax(Ytr[nn_index]), "true class:", np.argmax(Yte[i]))
        
        if np.argmax(Ytr[nn_index]) == np.argmax(Yte[i]):
            accuracy += 1.0 / len(Xte)
            
    print("done")
    print("accuracy: ", accuracy)
              
```
